## Investigation Overview


For the purpose this presentation, the Prosper loan dataset which contains 113,937 loans with 81 variables on each loan was collected, analyzed and visualized. The overall goal of this analysis was to uncover the factors drive loans to result in the various loan statuses: Completed, Defaulted, Chargedoff and Past Due and to also uncover the major reasons why borrowers apply for loans






_**The following insights were revealed through the analysis and visualization:**_
- There is a strong positive relationship between Income range and amount of loans granted. The average amount of loan granted is higher at higher income ranges and lower at lower income ranges. That is to say Borrower's income range affects the amount of loan to be granted to the borrower.
- Wedding, Cosmetic procedure and Business loans are likely to be defaulted and In general loans with interest rate higher than 20% are more likely to be defaulted while Borrowers are more likely to settle their Medical loans, Green loans, Large Purchases and RV
- The factors that most seemed to affect a loan's outcome was the amount of money borrowed and the interest rate of the loan. Higher interest, high value loans seems to be more likely to be past due. Employed individuals had the highest spread in the data, which makes sense since I assume employed individuals are more likely to apply for and receive loans.
- There is a strong negative relationship between borrowers credit score and interest rates on their loans and thus as the borrower's credit score increases, the interest rate decreases.

## Percentage Counts of Listing Categories

The analyst was interested in finding out the major reasons why borrowers apply for loans. The bar chart below depicts that Debt Consolidation accounted for 51.2%. That is to say more than half of the loans data were listed in Debt consolidation while Home Improvement and Business loans were ranked 4th and 5th respectively.

```python
def barplot():
    var_count = Prosper_Loan.ListingCategory.value_counts()
    n_var =  Prosper_Loan.ListingCategory.value_counts().sum()

    plt.figure(figsize=(14,8))
    order = Prosper_Loan.ListingCategory.value_counts().index
    color = sns.color_palette()[4]
    ax = sns.countplot(data=Prosper_Loan, y='ListingCategory', color = color, order = order)
    plt.xscale('log')
    ax.xaxis.set_major_formatter(ScalarFormatter())
    plt.xlabel('Count')
    plt.ylabel('Listing Catgory')
    plt.title('Percentage Counts of Listing Categories')

    # Logic to print the proportion text on the bars
    for i in range (var_count.shape[0]):

        count = var_count[i]

        # Convert count into a percentage, and then into string
        pct_string = '{:0.1f}%'.format(100*count/n_var)

        # Print the string value on the bar. 
        plt.text(count+1, i, pct_string, va='center');    
```

```python
barplot()
```

## Average Loan Amount per Income Range

As depicted by the Line plot, the average loan amount increases as the income range increases The averge loan amounts ranged from 4,273.97 to 13,073.13. Surprisingly, borrowers who earned 0 and those who were unemployed have a higher average Loan amounts than borrowers with an income range of 1-24,999

```python
def lineplot():
    OrderedIncomeRange = ['Not displayed', 'Not employed', '$0', '$1-24,999', '$25,000-49,999', '$50,000-74,999', 
                                    '$75,000-99,999', '$100,000+']
    IncomeCat = pd.api.types.CategoricalDtype(ordered=True, categories=OrderedIncomeRange)
    Prosper_Loan['IncomeRange'] = Prosper_Loan.IncomeRange.astype(IncomeCat)
    plt.figure(figsize=(14,8))
    color = sns.color_palette()[4]
    sns.pointplot(data=Prosper_Loan, x='IncomeRange', y='LoanOriginalAmount', color=color,
             ci='sd', linestyles="")
    plt.xlabel('Income Range')
    plt.ylabel('Average Loan Amount')
    plt.title('Average Loan Amount per Income Range');

    Prosper_Loan.groupby('IncomeRange')['LoanOriginalAmount'].mean()
```

```python
lineplot()
```

## Loan Status of Average Loan Amount per Listing Category

The category of the listing that the borrower selected when posting their listing and there are 21 different categories listed. The facet grid of bar plots below anlayses which of these categories are most likely to be completed, defaulted or charged off.
It is evident from the chart below that:
- Wedding loans, taxes and loans for cosmetic procedures are more likely to be defaulted than completed
- Green Loans, Engagement ring, Large purchase loans are more likely to be completed than defaulted



```python
def FacetGrid():
    status = ['Completed', 'Defaulted', 'Chargedoff']
    loan_data = Prosper_Loan.loc[Prosper_Loan['LoanStatus'].isin(status)]
    color = sns.color_palette()[4]

    g = sns.FacetGrid(data=loan_data, col='ListingCategory', col_wrap=4)
    g.fig.set_size_inches(14, 8)
    g.fig.suptitle('Loan Status of Average Loan Amount per Listing Category')
    g.map(sns.barplot, 'LoanStatus', 'LoanOriginalAmount', ci="sd", order=status, color=color)
    g.set_axis_labels('Loan Status','Avg L. Amt');    
```

```python
FacetGrid()
```
